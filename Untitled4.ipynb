```python
# 📁 Image Effect via KMeans Clustering - Google Colab Ready Version

import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from google.colab import files
from PIL import Image
from mpl_toolkits.mplot3d import Axes3D


# 📌 Upload Image
uploaded = files.upload()
filename = next(iter(uploaded))

# 📥 Load and Preprocess Image
img = cv2.imread(filename)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_resized = cv2.resize(img_rgb, (256, 256))

# 🧪 Reshape for KMeans
pixel_data = img_resized.reshape((-1, 3))

# 🔍 KMeans Clustering (adjust k for different effects)

k = 15  # You can change to 3, 7, 10, etc.
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixel_data)
labels = kmeans.labels_
palette = kmeans.cluster_centers_

# 🧱 Reconstruct Image from Cluster Centers
new_img = palette[labels].reshape(img_resized.shape).astype('uint8')

# 📊 3D and 2D Scatter Plot to Visualize RGB Clustering
sample_pixels = pixel_data[np.random.choice(pixel_data.shape[0], 1000, replace=False)]
fig = plt.figure(figsize=(16, 6))

# 3D RGB Scatter
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(
    sample_pixels[:, 0], sample_pixels[:, 1], sample_pixels[:, 2],
    color=sample_pixels / 255, s=10, alpha=0.6, label='Original Pixels'
)
ax1.scatter(
    palette[:, 0], palette[:, 1], palette[:, 2],
    color=palette / 255, s=200, marker='*', edgecolor='k', label='Centroids'
)
ax1.set_xlabel('Red')
ax1.set_ylabel('Green')
ax1.set_zlabel('Blue')
ax1.set_title(f'3D RGB Clustering - k={k}')
ax1.legend()

# 🎨 Show Only Original and KMeans Effect
plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.imshow(img_resized)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(new_img)
plt.title(f"KMeans Effect - K={k}")
plt.axis('off')

plt.tight_layout()
plt.show()
```
